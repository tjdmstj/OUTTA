```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver import ActionChains
```

```python
group = input('Input the artist name')
```

```python
driver = webdriver.Chrome()

driver.get('https://www.melon.com/')

search_box = driver.find_element(By.CSS_SELECTOR,'input.ui-autocomplete-input')
search_box.send_keys(group)
search_box.send_keys(Keys.RETURN)

artist = driver.find_element(By.CSS_SELECTOR,'#divCollection > ul > li:nth-child(3) > a')
#print(type(artist))
artist.click()

song_list = []
album_list = []
like_list = []
is_title = []
songs = driver.find_elements(By.CSS_SELECTOR,'#frm_defaultList > div > table > tbody >tr')

for song in songs:
    try:
        title = song.find_element(By.CSS_SELECTOR,'td:nth-child(3) > div > div > a.fc_gray')
        album = song.find_element(By.CSS_SELECTOR,'td:nth-child(5) > div > div > a')
        like = song.find_element(By.CSS_SELECTOR,'td:nth-child(6) > div > button > span.cnt')
        is_title_ = song.find_element(By.CSS_SELECTOR,'div.ellipsis > span')
        
        song_list.append(title.text)
        album_list.append(album.text)
        like_list.append(like.text)
        is_title.append(is_title_.text)
    except:
        pass
driver.quit()
```

```python
import pandas as pd
Artist = pd.DataFrame([is_title,song_list,album_list,like_list],index=['타이틀여부','곡명','앨범','좋아요']).T
Artist.head()
```

```python
def comma(like):
    try:
        heart = int(like.replace(",",""))
    except:
        heart = int(like)
    return heart
```

```python
Artist['좋아요'] = Artist['좋아요'].apply(comma)
Artist
```

```python
Artist[Artist['타이틀여부']=='Title'].sort_values(by='좋아요',ascending=False)
```
